Wczytaj do Pandas bazę najlepiej sprzedających się albumów wszechczasów.

```python
import pandas as pd
import numpy as np
import missingno as msno
```

```python
df = pd.read_html('https://www.officialcharts.com/chart-news/the-best-selling-albums-of-all-time-on-the-official-uk-chart__15551/', header=0)
df = df[0]
```

1. Zamień nagłówki kolumn na polskie odpowiedniki: ['TYTUŁ','ARTYSTA','ROK','MAX POZ']

```python
df.rename(columns={'POS':'LP', 'TITLE':'TYTUŁ',	'ARTIST':'ARTYSTA',	
                    'YEAR':'ROK',	'HIGH POSN':'MAX POZ'}, inplace=True)
```

```python
df
```

2. Ilu pojedynczych artystów znajduje się na liście?

```python
dfa=df['ARTYSTA'].nunique()
print(f'Na liście znajduje się {dfa} pojedynczych artystów.')
```

3. Które zespoły pojawiają się najczęściej na liście?

```python
dfva = df['ARTYSTA'].value_counts()
dfva = dfva.sort_values(ascending=False)
dfva = dfva[dfva>1].index
def dfva_a(dfva):
    dfv=[]
    for i in dfva:
        dfv.append(i)
    print(f'Najczęsciej na liście pojawiają sie: {dfv}')
dfva_a(dfva)
```

4. Zmień nagłówki kolumn, tak aby każdy z nich rozpoczynał się od wielkiej litery, a pozostałe były wprowadzone małymi literami.

```python
df.rename(str.title, axis='columns', inplace=True)
df
```

5. Wyrzuć z tabeli kolumnę ‘Max Poz’.

```python
df.drop(['Max Poz'], axis=1, inplace=True)
df
```

6. W którym roku wyszło najwięcej albumów znajdujących się na liście?

```python
dfgb = df.groupby('Rok').sum()
dfgb = dfgb.sort_values(by='Lp', ascending=False)
print(f'Najwięcej albumów znajdujących się na liście wyszło w {dfgb.index[0]} roku.')
```

7. Ile albumów wydanych między 1960 a 1990 rokiem włącznie znajduje się na liście?

```python
dfc = df[(df['Rok']>1960) & (df['Rok']<1991)]
dfc = (dfc['Rok']).count()
print(f'Na liście znajduje się {dfc} albumów wydanych Między 1960 a 1990 rokiem włącznie.')
```

8. W którym roku wydany został najmłodszy album na liście?

```python
dfrm = df['Rok'].min()
print(f'Najmłodszy album na liście został wydany w {dfrm} roku.')
```

9. Przygotuj listę najwcześniej wydanych albumów każdego artysty, który znalazł się na liście.

```python
dfs = df.sort_values(by='Rok')
dfs.drop_duplicates(['Artysta'],inplace=True)
dfs
```

10. Listę zapisz do pliku csv.

```python
dfs.to_csv('8_2_10_FirstAlbums.csv', index=False)
```
